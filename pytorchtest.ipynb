```python
import torch
import numpy as np
```

```python
data = [[1,2], [3,4]]
x_data = torch.tensor(data)
```

```python
np_arr = np.array(data)
x_np = torch.from_numpy(np_arr)
```

```python
x_ones = torch.ones_like(x_np)
x_rand = torch.rand_like(x_data, dtype = torch.float)
x_rand_int = torch.rand_like(x_rand, dtype = torch.float)
print(f"{x_ones}\n{x_rand}\n{x_rand_int}")
```

```python
shape = (2,9,)

rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print("{}\n{}\n{}".format(rand_tensor, ones_tensor, zeros_tensor))
```

```python
print("{} {} {}".format(rand_tensor.shape, rand_tensor.dtype, rand_tensor.device))
```

```python
tensor = torch.ones(4,4)
print(tensor)
print(f"row1 : {tensor[0]}")
print(f"col1 : {tensor[:, 0]}")

tensor[:, -1] = 0

print(f"""
      col_last : {tensor[:, -1]}
      col_last_alt : {tensor[..., -1]}
      
      notice : tensor[..., -1] = tensor[:,:,:,~~~생략~~~,-1]
""")
print(tensor)
```
